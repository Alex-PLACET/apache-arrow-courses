# Apache Arrow Course

Welcome to this hands-on course on **[Apache Arrow](https://arrow.apache.org)**, a cross-language development platform for in-memory analytics. This course covers the core concepts of Arrow's columnar memory format, its Python API (`pyarrow`), and key interoperability mechanisms.

## Course Outline

| # | Notebook | Description |
|---|----------|-------------|
| 01 | [Why Arrow?](01_why_arrow.ipynb) | Motivation and performance advantages of the Arrow format |
| 02 | [Arrays, Buffers & Nulls](02_arrays_buffers_nulls.ipynb) | Building blocks: typed arrays, memory buffers, and null handling |
| 03 | [Variable-length & Nested Types](03_variable_length_and_nested.ipynb) | Strings, lists, structs, and other complex types |
| 04 | [Dictionary & Run-End Encoding](04_dictionary_and_ree.ipynb) | Efficient encodings for repeated and run-length data |
| 05 | [Extension Types](05_extension_types.ipynb) | Defining custom semantic types on top of Arrow primitives |
| 06 | [C Data Interface: ctypes](06_c_data_interface_ctypes.ipynb) | Zero-copy data sharing across languages using ctypes |
| 07 | [C Data Interface: PyCapsule](07_c_data_interface_pycapsule.ipynb) | Zero-copy data sharing using the PyCapsule protocol |
| 08 | [IPC: Streaming & File Format](08_ipc_streaming_file.ipynb) | Serializing Arrow data with the IPC streaming and file formats |

## Prerequisites

- Basic Python knowledge
- Familiarity with tabular data (NumPy / pandas is helpful but not required)
